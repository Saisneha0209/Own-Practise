```python
#QUESTION NO.2:-
num = int(input("ENTER YOUR NUMBER:"))
divisor = int(input("ENTER DIVISOR:"))
if num%divisor==0:
  print(num,"is divided by",divisor,"!!")
else:
  print(num,"is not divided by",divisor,"!!")
```

```python
#QUESTION NO.4:-
num = int(input("enter a number:"))
for i in range(1,num+1):
  for j in range(1,i+1):
    print("*", end=" ")
  print()

```

```python
#question no.5:-
num =int(input("enter your number:"))
for i in range(1,num+1):
  for j in range(1,num-i+2):
    print(j,end = " ")
  print()
```
