```python
#File Handling:-
#FILE INPUT/OUTPUT!!
#OPEN,READ & CLOSE FILE!!
#r mmeans read and w means write!!
with open("LICENSE.txt","w") as file:
  file.write("Name:Saisneha")

with open("LICENSE.txt","r") as file:
  data = file.read()
  print(data)
```

```python
with open("LICENSE.txt","a") as file:
  file.write("Name:Saisneha")
  file.write("blaaablablaaaa")

with open("LICENSE.txt","r") as file:
  data = file.read(19)  #to read line by line !!  #or you can use readline
  print(data)
```

```python
#here in case of appending a new line on the last we can use 'a' instead of 'w' or 'r'!!
#DELETING A FILE:-
#We have to use a module to delete the file as we can not just simply delete it by writing delete.file!!
import os  #this is by default!!
os.remove("LICENSE.txt")
```

```python
def check_for_word():
  word="learning"
  with open("practise.txt","r") as file:
    data = file.read()
    if(data.find(word)!= -1):
      print("found")
    else:
      print("not found")

check_for_word()
```

```python
def check_for_line():
  word="pyq"
  data = True
  line_no=1
  with open("practise.txt","r") as file:
    while data:
      data = file.readline()
      if(word in data):
        print(line_no)
      line_no+=1
  return -1

print(check_for_line())
```
